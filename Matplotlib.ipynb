```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt 
```

MATPLOTLIB BASICS 

```python
x = np.linspace(0,5,11)
y = x**2
y
```

```python
plt.plot(x,y)
```

```python
plt.plot(x,y)
plt.title("Plot diagram using Matplotlib by Vials acharya ")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
```

SUBPLOT IN MATPLOTLIB

```python
plt.subplot(2,2,1)
plt.plot(x,y)
plt.subplot(2,2,4)
plt.plot(x,y)
plt.subplot(2,2,2)
plt.plot(x,y)
plt.subplot(2,2,3)
plt.plot(x**2,x**2)
```

```python
plt.subplot(6,4,1)
plt.subplot(6,4,2)
plt.subplot(6,4,3)
plt.subplot(6,4,4)
plt.subplot(6,4,5)
plt.subplot(6,4,6)
```

OBJECT ORIENTED WAY

```python
fig = plt.figure()
axis1=fig.add_axes([0.1,0.1,1,1])
axis1.plot(x,y)
axis2=fig.add_axes([0.3,0.3,0.3,1])
axis2.plot(x,y)
```

```python
fig= plt.figure()
ax=fig.add_axes([0.1,0.1,0.8,0.8])
ax.plot(x,y)
plt.savefig("Basicplot.png")
```
